```python
#  Import and read the USvideos.csv.
import pandas as pd 
USvideos_df= pd.read_csv("/Users/richellelong/Desktop/project/Reformatted_Trending_Dates.csv")
USvideos_df.head()
USvideos_df.dtypes
```

```python
# Drop other columns and keep only video_id, trending_date and publish_time.
Dates_df = USvideos_df.drop(["trending_date", "title", "channel_title", "category_id", "tags", "views", "likes", "dislikes", 
                        "comment_count", "thumbnail_link", "comments_disabled", "ratings_disabled", 
                        "video_error_or_removed", "description"], axis = 1)
Dates_df
```

```python
# identify datatypes
Dates_df.dtypes
```

```python
# format strings into desired datetime formats of YY-MM-DD
Dates_df["date_trending"] = pd.to_datetime(Dates_df["date_trending"]).dt.strftime('%Y-%m-%d')
Dates_df["publish_time"] = pd.to_datetime(Dates_df["publish_time"]).dt.strftime('%Y-%m-%d')
Dates_df.dtypes
Dates_df
```

```python
# format strings as datetime
Dates_df[["date_trending", "publish_time"]] = Dates_df[["date_trending", "publish_time"]].apply(pd.to_datetime)
# Dates_df
Dates_df.dtypes
```

```python
# Compute for difference between publish date and and trending_date
trending = Dates_df["date_trending"]
trending 

published = Dates_df["publish_time"]
published

difference_series = (trending - published)
difference_series
```

```python
# Merging the current dataframe with the series.
difference = difference_series
Dates_df['date_interval'] = difference_series
Dates_df
```

```python
# Save dataFrame as CSV
Dates_df.to_csv(r"/Users/richellelong/Desktop/project/Trending_Publish_Interval.csv", index = False)
```
