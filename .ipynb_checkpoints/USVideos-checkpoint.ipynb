```python
# load Trending_Publish_Interval.csv
dates_df = pd.read_csv("ETL/Intervals.csv")
dates_df
```

```python
# Concatonate us_videos with dates_df 
df_videos_US = pd.concat([usvideos_df, dates_df], axis=1) 
df_videos_US
```

```python
df_videos_US.dtypes
```

```python
df_videos_US.drop(['trending_date','publish_time', 'video_id_X'], axis = 1, inplace = True)
df_videos_US
```

```python
column_headers = ['video_id', 'date_trending', 'date_interval', 'title', 'channel_title', 'category_id',
       'date_published', 'tags', 'views', 'likes', 'dislikes', 'comment_count',
       'thumbnail_link', 'comments_disabled','ratings_disabled',
       'video_error_or_removed']
new_df = df_videos_US.reindex(columns=column_headers)
new_df
```

```python
# format strings as datetime
new_df[["date_trending", "date_published"]] = new_df[["date_trending", "date_published"]].apply(pd.to_datetime)
# Dates_df
new_df.dtypes
```

```python
# Renaming dataFrame before removing duplicates.
new_usvideos_df = new_df
```

# Part B ~ Create dataframe according to number of views in descending order

```python
# sort views column descending order
new_usvideos_df.sort_values(by=['views'], ascending=False)
```

```python
# Rename DataFrame that is arranged according to number of views in descending order.
views_descending = new_usvideos_df.sort_values(by=['views'], ascending=False)
```

```python
# Count duplicate in a DataFrame
views_descending.video_id.duplicated().sum()
```

```python
views_descending.dtypes
```

```python
# Saved descending views as CSV
# views_descending.to_csv("Resources/ETL2_desc_order.csv", index = False)
```

# Part C ~ first trending date (deleted duplicate rows)

```python
# group data over columns 'video_id' and 'date_published' and keep last entry only
first_trending_date = views_descending.groupby(['title']).last()
  
# print new dataframe
first_trending_date
```

```python
# Saved dataframe with first trending date
first_trending_date.to_csv("Resources/ETL3_first_tdate.csv", index = False)
```

# Part D ~ last trending date (deleted duplicate rows)

```python
# group data over columns 'video_id' and 'date_published' and keep first entry only
last_trending_date = views_descending.groupby(['title']).first()
  
# print new dataframe
last_trending_date
```

```python
# Saved dataframe in CSV
last_trending_date.to_csv("Resources/ETL4_last_tdate.csv", index = False)
```
